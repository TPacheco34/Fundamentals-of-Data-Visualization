```python
import altair as alt
import pandas as pd
from vega_datasets import data
```

```python

housing_data = pd.read_csv('Documents/Masters_Program/Visualization/housing_gowth_metro.csv')

housing_data_zip = pd.read_csv('Documents/Masters_Program/Visualization/housing_growth_zipcode.csv')
```

```python
housing_data.head()

housing_data = housing_data.rename(columns = {'RegionName':'cities'})

metro_data = housing_data.filter(['cities','2023-08-31'], axis = 1)





metro_data_max = metro_data.sort_values(by = ['2023-08-31'], ascending = False)

metro_data_max.drop(metro_data_max.index[15:], inplace = True)




metro_data_min = metro_data.sort_values(by = ['2023-08-31'], ascending = True)

metro_data_min.drop(metro_data_min.index[15:], inplace = True)

metro_data_min_max = metro_data.sort_values(by = ['2023-08-31'], ascending = False)

metro_data_min_max.drop(metro_data_min_max.index[30:869], inplace = True)



metro_change = housing_data.filter(['cities', '2022-09-30', '2022-11-30', '2023-08-31'], axis = 1)



zip_data = housing_data_zip.filter(['CountyName','2023-08-31'], axis = 1)
```

```python
metro_data_change = housing_data.filter(['cities','2022-09-30','2023-08-31'], axis = 1)

metro_data_change['Year Change'] = metro_data_change['2023-08-31'] - metro_data_change['2022-09-30']
```

```python
metro_data_change.head()
```

```python
source = metro_data_change

base = alt.Chart(source).encode(
    alt.X('cities:N', axis=alt.Axis(title= None))
)

point1 = base.mark_point(color='#57A44C').encode(
    alt.Y('2022-09-30',
          axis=alt.Axis(title='Growth'))
)

point2 = base.mark_point(interpolate='monotone').encode(
    alt.Y('2023-08-31',
          axis=alt.Axis(title= 'Growth'))
)

alt.layer(point1, point2)
```
